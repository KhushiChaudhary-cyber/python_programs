```python
# Import matplotlib library
import matplotlib.pyplot as plt

# Months data
months = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct"]

# Rainfall data in millimeters
rainfall = [10,15,20,30,40,80,120,100,60,30]

# Plot rainfall data
plt.plot(months, rainfall, marker='o')

# Chart title
plt.title("Monthly Rainfall(mm)")

# X-axis label
plt.xlabel("Month")

# Y-axis label
plt.ylabel("Rainfall (mm)")

# Display the chart
plt.show()

```
